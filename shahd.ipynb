```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\user\Desktop\ObesityDataSet_raw_and_data_sinthetic.csv")
```

```python
dataset.info()
```

```python
#Finding Numerical Columns
num_col=[]
for col in dataset.columns:
    if(dataset[col].dtypes!='object'):
        num_col.append(col)
print(f"There are total {len(num_col)} numerical columns in dataset")
print(num_col)
```

```python
#Finding Categorical Columns
cat_col=[]
for col in dataset.columns:
    if(dataset[col].dtypes=='object'):
        cat_col.append(col)
print(f"There are total {len(cat_col)} categorical columns in dataset")
print(cat_col)
```

```python
data=dataset.copy(deep=True)
```

```python
data[cat_col].sample(5)
```

```python
data[num_col].sample(5)
```

```python
#"How often do you have physical activity?"


pd.pivot_table(data,index=['NObeyesdad'],columns='Gender',values='FAF',margins=True).T.plot(kind='bar')
```

```python
#Height and Weight

pd.pivot_table(data,index='Gender',columns='NObeyesdad',values=['Height','Weight'],margins=True).T.plot(kind='bar')
```

```python
#"How much water do you drink daily?"

pd.pivot_table(data,index='Gender',columns='NObeyesdad',values=['CH2O'],margins=True).T.plot(kind='bar')
```

```python
#Age

pd.pivot_table(data,index='Gender',columns='NObeyesdad',values=['Age'],margins=True).T.plot(kind='bar')
```

```python
#"Do you usually eat vegetables in your meals?"

pd.pivot_table(data,index='Gender',columns='NObeyesdad',values=['FCVC'],margins=True).T.plot(kind='bar')
```

```python
#"How many main meals do you have daily? "
pd.pivot_table(data,index='Gender',columns='NObeyesdad',values=['NCP'],margins=True).T.plot(kind='bar')
```
